```python
import os

from PIL import Image
from IPython.display import display, Image as IPImage
```

```python
# Define paths to the original and generated image folders
original_path = '/content/drive/MyDrive/CS245 Project/x-ray-edges-json/Test/target'
generated_path_nofinetune = '/content/drive/MyDrive/CS245 Project/ControlNet_UnFinetune_TestOutput'
generated_path_finetune_5 = '/content/drive/MyDrive/CS245 Project/ControlNet_Finetune_TestOutput_5'
generated_path_finetune_25 = '/content/drive/MyDrive/CS245 Project/ControlNet_Finetune_TestOutput_25'
generated_path_finetune_50 = '/content/drive/MyDrive/CS245 Project/ControlNet_Finetune_TestOutput_50'

generated_paths = [
    generated_path_nofinetune,
    generated_path_finetune_5,
    generated_path_finetune_25,
    generated_path_finetune_50
]
```

# Pairwise Visualization

```python
# Customize display size
display_size = (200, 200)  # Adjust the size as needed
```

```python
# Load and preprocess images
def load_and_preprocess_images(path):
    return [Image.open(os.path.join(path, file)).resize(display_size) for file in sorted(os.listdir(path))]

# Get the list of filenames in each folder
original_files = load_and_preprocess_images(original_path)
generated_files = [load_and_preprocess_images(path) for path in generated_paths]

# Display images with titles
for i, (original, *generated) in enumerate(zip(original_files, *generated_files)):
    images_to_display = [IPImage(img._repr_png_()) for img in [original, *generated]]
    display(*images_to_display)

    if i >= 0:
        break
```

# Visualize one example


```python
def find_and_load_images(filename, display_size):
    images = [Image.open(os.path.join(path, filename)).resize(display_size) for path in generated_paths]
    return tuple(images)

# Filename to search for
search_filename = '00004523_003.png'

# Find and load images with the specified filename
original_img, *finetune_images = find_and_load_images(search_filename, display_size)

# Display images with titles
display(original_img, *finetune_images)

# Print filenames
print(f"{search_filename}")
```
